# Plotting different kind of plots in Seaborn

# Histogram

```python
# sns.histplot()
# sns.displot()
```

```python
# sns.histplot(numerical column)

# stat=density,probability,count, frequency
```

```python
plt.figure(figsize=(8,6),dpi=150)
sns.histplot(x='age',data=tt,bins=30,stat='density',color='red')

plt.show()
```

```python
plt.figure(figsize=(6,4),dpi=100)
sns.histplot(tt.age,kde=True,color='red')

plt.show()
```

```python
# kde- kernel density estimation
```

```python
# sns.displot(numerical column)

# kind=hist/kde
```

```python
sns.displot(tt.age,color='green',kind='hist',height=8,aspect=1.5)

plt.show()
```

# KDE

```python
# sns.kdeplot(numerical column)
```

```python
sns.kdeplot(tt.age)

plt.show()
```

```python
sns.kdeplot(x='age',data=tt,color='green',fill=True)

plt.show()
```

# Boxplot

```python
# sns.boxplot(numerical column)

# sns.boxplot(x=Numerical column,y=Categorical,hue='categorical',data=df)
```

```python
sns.boxplot(y='age',x='pclass',hue='sex',data=tt)

plt.show()
```

```python
plt.figure(figsize=(8,6),dpi=100)
sns.boxplot(y='age',x='sex',hue='pclass',data=tt)

plt.legend(loc=10)
plt.show()
```
