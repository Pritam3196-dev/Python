# Giving the title and the axes names

```python
# plt.title('title')- Title
# plt.xlabel('axis-name')- Name for X- Axis
# plt.ylabel('axis-name')- Name for Y- Axis
```

```python
plt.plot(arr,arr_2)
plt.title('My plot',fontdict={'color':'red','font':'cursive','size':15})
plt.xlabel('Array-1',fontdict={'color':'red','font':'cursive','size':15})
plt.ylabel('Array-2',fontdict={'color':'red','font':'cursive','size':15})

plt.show()
```

# Changing the line color, line style etc. and giving markers

```python
# color
# line style-  ls -> -,--,-.,:
# Line width- lw -> int/float
# marker=,'o','v','*','p',1,2,3,4,8,'<','>'
```

```python
plt.plot(arr,arr_2,color='red',ls='-',lw=1,marker='o',markerfacecolor='yellow',markeredgecolor='blue',markersize=12,markeredgewidth=2)

plt.title('My plot',fontdict={'color':'red','font':'cursive','size':15})
plt.xlabel('Array-1',fontdict={'color':'red','font':'cursive','size':15})
plt.ylabel('Array-2',fontdict={'color':'red','font':'cursive','size':15})

plt.show()
```

# Giving Legend

```python
plt.plot(arr,arr_2,marker='o',markersize=12,color='r',label='Trend')

plt.legend(loc=10,fontsize=15,labelcolor='blue')

plt.show()
```

# Object Oriented Approch

```python
# - It is also known as figure and axes approach.
# - It is more flexible than the functional approach.
# - Using this we can manipulate the sizes of the plots
```

```python
# axis- Singular
# axes- Plural
```

# Creating a figure

```python
f=plt.figure()
```

# Creating axes in a figure

```python
# figure_object.add_axes([left_boundary,bottom_boundary,width,height])
```

```python
f=plt.figure()

a=f.add_axes([0.1,0.1,0.5,0.5])
b=f.add_axes([0.7,0.1,0.5,0.5])
c=f.add_axes([0.7,0.7,0.8,0.5])
d=f.add_axes([0.1,0.7,0.5,0.7])

a.plot(arr,arr_2)
b.plot(df.A,df.E)
c.plot(df.B,df.D)
d.plot(df.C,df.A**2)


plt.show()
```

# Giving title and axes names

```python
# axes_name.set_title('title')
# axes_name.set_xlabel('X-Axis_name')
# axes_name.set_ylabel('Y-Axis-name')
```

```python
f=plt.figure(figsize=(10,8),dpi=100)

a=f.add_axes([0.1,0.1,0.5,0.5])
b=f.add_axes([0.7,0.1,0.5,0.5])
c=f.add_axes([0.7,0.7,0.8,0.5])
d=f.add_axes([0.1,0.7,0.5,0.7])

a.plot(arr,arr_2)
a.set_title('Plot-A')
a.set_xlabel('X-Axis')
a.set_ylabel('Y-Axis')

b.plot(df.A,df.E)
b.set_title('Plot-B')
c.plot(df.B,df.D)
d.plot(df.C,df.A**2)


plt.show()
```

```python
f.savefig('Plot_A.jpeg')
plt.show()
```

# plt.subplots()

```python
# - It is an automatic axes manager.
# - It creates a figure and axes both
# - It returns a tuple of two elements.
# - The first element in the tuple is the figure.
# - The second element in the tuple is a numpy array.
# - The array contains the axes objects
```

```python
# plt.subplots(rows,cols)
```

```python
f,a=plt.subplots(1,2,figsize=(8,4),dpi=100)

a[0].plot(df.A,df.B)

a[1].plot(df.B,df.A)

plt.show()
```
