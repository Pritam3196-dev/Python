# Iterable

```python
# - Any object which we can apply for loop upon is an iterable.
# - Any object that has __iter__ method defined in its class is an iterable.

# Examples- string, list,tuple, dictionary, set, frozenset, range etc.
```

```python
a='Python'

for i in a:
    print(i)
```

```python
a=10

for i in a:
    print(i)
```

```python
a='Python'

print(dir(a))
```

```python
a=10

print(dir(a))
```

# Iteration

```python
# - Traversing upon the elements of an iterable one by one from the beginning to the end is known as iteration
```

```python
a='Python'
```

```python
for i in [10,20,30,40]:
    print(i)
```

```python
# i=10
# i=20
# i=30
# i=40
```

```python
# Iterable -> fetching the elements one by one in sequence -> This is not possible
```

```python
# Iterable -> Iterator -> Fetch the elements one by one 
```

```python
# - We can not fetch the elments from an iterable one by one.
# - We have to convert the iterable into iterator.
# - From the iterator we can fetch the elements one by one
```

# Converting an iterable into an iterator

### iter()

```python
# - It takes an iterable as an argument and returns an iterator object.
```

```python
a='Python'

x=iter(a)
print(x,type(x))
```

```python
lst=[10,20,30,40]

y=iter(lst)
print(y,type(y))
```
